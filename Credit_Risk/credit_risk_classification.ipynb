```python
# Import the modules
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.metrics import confusion_matrix, classification_report,accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from imblearn.over_sampling import RandomOverSampler
from imblearn.ensemble import BalancedRandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
import warnings
# Ignore all warnings
warnings.filterwarnings('ignore')
```

## Split the Data into Training and Testing Sets

### Step 1: Read the `lending_data.csv` data from the `Resources` folder into a Pandas DataFrame.

```python
# Read the CSV file from the Resources folder into a Pandas DataFrame
lending_df = pd.read_csv("Resources/lending_data.csv")


# Review the DataFrame
lending_df.head()
```

```python
#Value count of loan status
lending_df["loan_status"].value_counts()
```

### Step 2: Create the labels set (`y`)  from the “loan_status” column, and then create the features (`X`) DataFrame from the remaining columns.

```python
# Separate the data into labels and features
# Separate the y variable, the labels
target=lending_df["loan_status"]
# Separate the X variable, the features
data = lending_df.drop("loan_status",axis=1)
feautres_df=data.columns
```

```python
# Review the y variable Series
target
```

```python
# Review the X variable DataFrame
feautres_df
```

### Step 3: Split the data into training and testing datasets by using `train_test_split`.

```python
# Import the train_test_learn module
#from sklearn.model_selection import train_test_split

# Split the data using train_test_split
# Assign a random_state of 1 to the function
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=1)
```

```python
#value of count of training dataset
#X_train.shape
train_data=pd.Series(y_train).value_counts()
train_data
```

```python
#value of count of testing dataset
#x_test.shape()
test_data=pd.Series(y_test).value_counts()
test_data
```

## Create a Logistic Regression Model with the Original Data

###  Step 1: Fit a logistic regression model by using the training data (`X_train` and `y_train`).

```python
# Import the LogisticRegression module from SKLearn
#from sklearn.linear_model import LogisticRegression

# Instantiate the Logistic Regression model
# Assign a random_state parameter of 1 to the model
classifier = LogisticRegression(solver='lbfgs', random_state=1)
classifier

# Fit the model using training data
classifier.fit(X_train, y_train)
```

```python
#Make a prediction using train data
predictions_train = classifier.predict(X_train)
```

### Step 2: Save the predictions on the testing data labels by using the testing feature data (`X_test`) and the fitted model.

```python
# Make a prediction using the testing data
predictions_test = classifier.predict(X_test)
pd.DataFrame({"Prediction": predictions_test, "Actual": y_test})
```

```python
#accuracy score for the train data
acc_score_test = accuracy_score(y_test, predictions_test)
print(f"accuracy of test data:{acc_score_test}")
```

```python
#accuracy score for the train data
acc_score_train = accuracy_score(y_train, predictions_train)
print(f"accuracy of train data:{acc_score_train}")
```

### Step 3: Evaluate the model’s performance by doing the following:

* Generate a confusion matrix.

* Print the classification report.

```python
# Generate a confusion matrix for the model
cm=confusion_matrix(y_test, predictions_test)
cm_df=pd.DataFrame(cm,index=["healthyloan(0)","highriskloan(1)"],columns=["predicted(0)","predicted(1)"])
cm_df
```

```python
#confusion_matrix(y_test, predictions)
```

```python
# Print the classification report for the model

print("Logistic Regression")
target_names = ["healthy", "highrisk"]
print(f"Accuracy Score : {acc_score_test}")
print(classification_report(y_test, predictions_test, target_names=target_names))
```

### Random Over sampler with Logistic regression 

```python
#install imbalanced-learn
#!pip install imbalanced-learn
```

```python
#instantiate the random_oversampler 
Sampler = RandomOverSampler(random_state=1)
X_sample, y_sample = Sampler.fit_resample(X_train,y_train)

# Fit the original training data to the random_oversampler 
X_train, X_test, y_train, y_test = train_test_split(X_sample, y_sample, random_state=1)
```

```python
# Instantiate the Logistic Regression model
# Assign a random_state parameter of 1 to the model
classifier = LogisticRegression(solver='lbfgs', random_state=1)
classifier

# Fit the model using training data
classifier.fit(X_train, y_train)
```

```python
# Make a prediction using the testing data
predictions_test_res = classifier.predict(X_test)
pd.DataFrame({"Prediction": predictions_test_res, "Actual": y_test})
```

```python
#accuracy score using Randomoversampler
predictions_train_res = classifier.predict(X_train)
acc_score_train = accuracy_score(y_train, predictions_train_res)
print(f"accuracy score of train data usingrandom oversampler:{acc_score_train}")
```

```python
#accuracy score using Randomoversampler
acc_score_test=accuracy_score(y_test, predictions_test_res)
print(f"accuracy  score of test data using  random oversampler:{acc_score_test}")
```

---

```python
#Generate confusion matrix
cm=confusion_matrix(y_test, predictions_test_res)
cm_df=pd.DataFrame(cm,index=["healthyloan(0)","highriskloan(1)"],columns=["predicted(0)","predicted(1)"])
cm_df
```

```python
# Print the classification report for the model

print("Logistic Regression using RandomOverSampler")
target_names = ["healthy", "highrisk"]
print(f"Accuracy Score : {acc_score_test}")
print(classification_report(y_test, predictions_test_res, target_names=target_names))
```

###   RandomOverSampler with Balanced Random Forest 

```python
#instantiate the random_oversampler 
Sampler = RandomOverSampler(random_state=1)
X_rf, y_rf = Sampler.fit_resample(X_train,y_train)

# Fit the original training data to the random_oversampler 
X_train, X_test, y_train, y_test = train_test_split(X_rf, y_rf, random_state=1)
```

```python

# Create a Balanced Random Forest classifier
rf = BalancedRandomForestClassifier(n_estimators=500, random_state=1)

# Train the classifier
rf.fit(X_train, y_train)




```

```python
#Make predictions on the test set
prediction_rf = rf.predict(X_test)
```

```python
#Generate confusion matrix
cm=confusion_matrix(y_test, prediction_rf)
cm_df=pd.DataFrame(cm,index=["healthyloan(0)","highriskloan(1)"],columns=["predicted(0)","predicted(1)"])
cm_df
```

```python
# print the classification report
print("Balanced Random Forest using random over sampler")
print(classification_report(y_test, prediction_rf))
```

###  RandomOverSampler with  Support Vector Machines

```python
#instantiate the random_oversampler 
Sampler = RandomOverSampler(random_state=1)
X_svm, y_svm = Sampler.fit_resample(X_train,y_train)

# Fit the original training data to the random_oversampler 
X_train, X_test, y_train, y_test = train_test_split(X_svm, y_svm, random_state=1)
```

```python

# Create a svm classifier
svm = SVC(random_state=1)

# Train the classifier
svm.fit(X_train, y_train)


```

```python
# Make predictions on the test set
prediction_svm = svm .predict(X_test)
```

```python
#Generate confusion matrix
cm=confusion_matrix(y_test, prediction_svm)
cm_df=pd.DataFrame(cm,index=["healthyloan(0)","highriskloan(1)"],columns=["predicted(0)","predicted(1)"])
cm_df
```

```python
# print the classification report
print("Suppor Vector Machines using random over sampler")
print(classification_report(y_test, prediction_svm))
```
